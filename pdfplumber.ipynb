```python
import pdfplumber
import re
# 00010400.pdf
# 99123101.pdf
file = "D:/HEINRICH/Projects/Scrapy-app/data/23030200.pdf"
```

```python
import pdfplumber

result = []

with pdfplumber.open(file) as pdf:
    for page in pdf.pages:
        # Extract all words on each page
        raw_words = page.extract_text()
        
        print(raw_words)
                
        # Remove the $ and . and , symbols
        pattern = r"[$\.,]+"
        data = [re.sub(pattern, "", x) for x in raw_words.split("\n")]
        
        for i in range(len(data)):
            text = data[i].split(" ")
            text = list(filter(None, text))
            
            print(text)
            

            

```
